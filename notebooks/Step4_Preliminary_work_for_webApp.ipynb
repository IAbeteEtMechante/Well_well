## Code to launch at the start of the webApp

We digitalized the ESRI fault map of Nevada into 3 parts

```python
import numpy as np
import matplotlib.pyplot as plt
import PIL
from PIL import Image
from numpy import asarray
%matplotlib inline
```

```python
#load images of faults
img1 = Image.open('ESRI_Nevada_North.png').convert("RGB")
img2 = Image.open('ESRI_Nevada_Central.png').convert("RGB")
img3 = Image.open('ESRI_Nevada_South.png').convert("RGB")
img1 = asarray(img1)
img2 = asarray(img2)
img3 = asarray(img3)
# print(img1.shape)
# print(img2.shape)
# print(img3.shape)
```

```python
#combine all images in one big image of whole Nevada
img = np.concatenate((img1, img2), axis=0)
img = np.concatenate((img, img3), axis=0)
# print(img.shape)
```

```python
#using https://imagecolorpicker.com to get the colors from the ESRI legend
colors = np.zeros((7,3))
colors[0] = [255,147,142]
colors[1] = [255,177,151]
colors[2] = [242,220,161]
colors[3] = [166,187,138]
colors[4] = [136,166,211]
colors[5] = [128,128,128]
colors[6] = [167,167,167]

colors
```

```python
#modify the Nevada faults image to have only the faults as simple lines and nothing else
height, width, _ = img.shape
modified_img = img.copy()
for i in range(height):
  if i % 100 == 0:
    print(i)
  for j in range(width):
    for color in colors:
      if np.linalg.norm(color - img[i,j]) < 20:
        modified_img[i,j] = color
        break
    else:
      modified_img[i,j] = [255,255,255]
```

```python
np.save('Nevada_faults.npy', img)    # .npy extension is added if not given
```
